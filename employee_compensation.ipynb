```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
data_scaled = scaler.fit_transform(df1)
```

```python
data_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(data_scaled)
```

```python
kmeans.cluster_centers_
```

```python
y = kmeans.predict(data_scaled)
```

```python
y
```

```python
from scipy.spatial.distance import cdist, pdist
from sklearn.cluster import KMeans
```

```python
sum_squared_distance=[]
k=range(1,11)
for n_clusters in k:
    km=KMeans(n_clusters=n_clusters,init="k-means++",n_init=10,max_iter=300,random_state=1)
    km.fit(data_scaled)
    sum_squared_distance.append(km.inertia_)
```

```python
plt.plot(k,sum_squared_distance)
plt.xlabel('k')
plt.ylabel('sum_squared_distance')
plt.title('Elbow curve of KMeans')
plt.show()
```

```python
sns.countplot(x = kmeans.labels_, palette = "inferno")
```

```python
from sklearn.metrics import silhouette_samples,silhouette_score
silhouette_score(data_scaled, y)
```
